```python
import os
import json
import re
from langchain.document_loaders import PyPDFLoader

# í´ë” ê²½ë¡œ ì„¤ì •
DATA_PATH = "./data"  # PDF íŒŒì¼ í´ë”
OUTPUT_JSON = "./cleaned/cleaned_data.json"

# í…ìŠ¤íŠ¸ ì •ì œ í•¨ìˆ˜
def clean_text(text):
    text = re.sub(r"í˜ì´ì§€\s*\d+\s*/\s*\d+", "", text)
    text = re.sub(r"â€».*", "", text)
    text = re.sub(r"\s{2,}", " ", text)
    return text.strip()

# PDF ë¡œë“œ ë° ì •ì œ -> JSON ì €ì¥
def extract_cleaned_data():
    cleaned_docs = []
    os.makedirs(os.path.dirname(OUTPUT_JSON), exist_ok=True)

    for root, _, files in os.walk(DATA_PATH):
        for file in files:
            if file.endswith(".pdf"):
                pdf_path = os.path.join(root, file)
                print(f"ğŸ“„ ë¡œë”© ì¤‘: {pdf_path}")
                loader = PyPDFLoader(pdf_path)

                try:
                    docs = loader.load()
                    for doc in docs:
                        cleaned = clean_text(doc.page_content)
                        if cleaned:
                            cleaned_docs.append({
                                "source": pdf_path,
                                "content": cleaned
                            })
                except Exception as e:
                    print(f"âŒ ì‹¤íŒ¨: {file} â†’ {e}")

    with open(OUTPUT_JSON, "w", encoding="utf-8") as f:
        json.dump(cleaned_docs, f, ensure_ascii=False, indent=2)

    print(f"âœ… ì´ {len(cleaned_docs)}ê°œ í˜ì´ì§€ ì €ì¥ ì™„ë£Œ â†’ {OUTPUT_JSON}")

# ì‹¤í–‰
extract_cleaned_data()
```
